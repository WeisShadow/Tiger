```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Parmetros del dominio
x_max = 50.0
y_max = 50.0
nx, ny = 51, 51
x = np.linspace(0, x_max, nx)
y = np.linspace(0, y_max, ny)
X, Y = np.meshgrid(x, y)

# 2. Inicializacin de la matriz phi (potencial)
phi = np.zeros((ny, nx))

# 3. Aplicacin de Condiciones de Frontera
# Izquierda (x = 0) y Derecha (x = x_max)
phi[:, 0] = 4
phi[:, -1] = 9

# Lado inferior (y = 0) -> Rampa lineal
for i in range(nx):
    phi[0, i] = 4 + (9 - 4) * (x[i] / x_max)

# Lado superior (y = y_max) -> Funcin seccionada
for i in range(nx):
    if x[i] < x_max / 2:
        phi[-1, i] = 4 + 2 * (12 - 4) * (x[i] / x_max)
    else:
        phi[-1, i] = 15 + 2 * (9 - 12) * (x[i] / x_max)

# 4. Mtodo Iterativo (Teorema del Valor Medio con 8 vecinos)
tol = 1e-4
max_iter = 5000
error = 1.0
iteration = 0

while error > tol and iteration < max_iter:
    phi_old = phi.copy()

    # El promedio se hace sobre los puntos internos (1 a n-1)
    for j in range(1, ny - 1):
        for i in range(1, nx - 1):
            # Suma de los 8 vecinos (Cardinales + Diagonales)
            sum_8_neighbors = (
                phi_old[j+1, i] + phi_old[j-1, i] +   # Arriba, Abajo
                phi_old[j, i+1] + phi_old[j, i-1] +   # Derecha, Izquierda
                phi_old[j+1, i+1] + phi_old[j-1, i-1] + # Diagonales principales
                phi_old[j+1, i-1] + phi_old[j-1, i+1]   # Diagonales secundarias
            )
            phi[j, i] = sum_8_neighbors / 8.0

    error = np.max(np.abs(phi - phi_old))
    iteration += 1

print(f"Convergencia alcanzada en {iteration} iteraciones con error {error:.2e}")

# 5. Visualizacin
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, phi, cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Potencial $\phi(x,y)$')
ax.set_title('Solucin a la Ecuacin de Laplace (8 Vecinos)')
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
