# **Problema 1**

**Caída de una pelota desde una torre:** Se deja caer una pelota desde una torre de altura h. Tiene velocidad inicial cero y acelera hacia abajo con la gravedad.

Escribe un programa que le pida al usuario que ingrese la altura en metros de la torre y luego calcule e imprima el tiempo en segundos hasta que la pelota toque el suelo (ignora la resistencia del aire).

Después, usa tu programa para calcular el tiempo de una pelota lanzada desde una
torre de 100m de altura.


```python
import math
altura = float(input("Ingresa la altura en metros: "))
gravedad = 9.81
tiempo = math.sqrt((2 * altura) / gravedad)
print(f"El tiempo que tarda la pelota en tocar el suelo es de {tiempo:.2f} segundos. ")
```

# **Problema** **3**
**Coordenadas polares y cartesianas**: Supongamos que se nos da (es decir, el usuario introduce) la posicion de un punto en un espacio bidimensional:  ́

**a)** Si el punto se especifica en coordenadas polares r,θ. Escribe un programa (en
python sin usar la biblioteca numpy) que calcule e imprima la conversion a coordenadas cartesianas x, y.

```python
import math

def polares(r,theta):

  x = r * math.cos(math.radians(theta))
  y = r * math.sin(math.radians(theta))

  return x,y

r = float(input("valor de r"))
theta = float(input("valor de theta (radianes)"))

x,y = polares(r, theta)
print(f"Coordenadas cartesianas: x = {x}, y = {y}")
```

**b)** Si el punto se especifica en coordenadas cartesianas x, y. Escribe un programa (en python sin usar numpy) que calcule e imprima la conversion a coordenadas polares r,θ.

```python
import math

def cartesianas(u,v):
  d = math.sqrt(u**2+v**2)
  alpha = math.atan2(v,u)
  return d,alpha


u = float(input("valor de eje horizontal"))
v = float(input("valor del eje vertical"))

d,alpha = cartesianas(u,v)
print(f"Coordenadas polares r = {d} , theta = {alpha} (rad)")
```

# **Problema 4**

a) La segunda ley de Kepler nos dice que la distancia l2 y la velocidad v2 del planeta
en su punto más distante, o afelio, satisfacen que l2v2 = l1v1. Al mismo tiempo,
la energía total, cinética más la gravitatoria, de un planeta con velocidad v y distancia r del Sol está dada por:

E=(1/2)(mv2) − G mM/r

donde m es la masa del planeta, M= 1,9891 × 1030 kg es la masa del Sol y G=6,6738 × 10−11 m3 kg−1 s−2 es la constante gravitatoria de Newton.

Dado que la energía debe conservarse, demuestre que v2 es la raíz más pequeña de la ecuación cuadrática:

v2^2 - (2GMv2)/(v1l1)-[v1^2 - (2GM/l1] = 0.

Una vez que tenemos v2 podemos calcular l2 usando la relación l2 = l1v1/v2.

b) Dados los valores de v1, l1, y l2; otros parámetros de la órbita se obtienen mediante fórmulas simples que pueden derivarse de las leyes de Kepler y del hecho
de que la órbita es una elipse:

Semieje mayor: a = (1/2) (l1 + l2),

Semieje menor: b= (l1l2) ,

Período orbital: T=2πab / (l1v1)

Excentricidad orbital: e = (l2 - l1)/(l2 + l1).

Escribe un programa que le pida al usuario que ingrese la distancia al Sol y la velocidad en el perihelio; para que calcule e imprima las cantidades l2, v2, T y e.

c) Prueba tu programa haciendo que calcule las propiedades de las órbitas de la Tierra (para las cuales l1 = 1,4710 × 1011 m y v1 = 3,0287 × 104 m s−1) y del cometa Halley (l1 = 8,7830 × 1010 m y v1 = 5,4529 × 104 m s−1). Entre otras cosas, deberías encontrar que el período orbital de la Tierra es de un año y el del cometa Halley es de unos 76 años.

```python
import math

def calcular_parametros_orbitales(l1, v1):
    G = 6.6738e-11  # Constante gravitacional (m^3 kg^-1 s^-2)
    M = 1.9891e30   # Masa del Sol (kg)

    # Resolver ecuación cuadrática para v2
    a = 1
    b = - (2 * G * M) / (v1 * l1)
    c = - (v1 ** 2 - (2 * G * M) / l1)

    discriminante = b**2 - 4 * a * c
    if discriminante < 0:
        raise ValueError("No hay solución real para v2")

    v2 = (-b - math.sqrt(discriminante)) / (2 * a)  # Se elige la raíz más pequeña

    # Calcular l2
    l2 = (l1 * v1) / v2

    # Calcular otros parámetros
    a = 0.5 * (l1 + l2)  # Semieje mayor
    b = math.sqrt(l1 * l2)  # Semieje menor
    T = (2 * math.pi * a * b) / (l1 * v1)  # Período orbital
    e = (l2 - l1) / (l2 + l1)  # Excentricidad orbital

    return l2, v2, T, e

# Solicitar datos al usuario
l1 = float(input("Ingrese la distancia al Sol en el perihelio (m): "))
v1 = float(input("Ingrese la velocidad en el perihelio (m/s): "))

l2, v2, T, e = calcular_parametros_orbitales(l1, v1)

# Mostrar resultados
print(f"Distancia en el afelio: {l2:.3e} m")
print(f"Velocidad en el afelio: {v2:.3e} m/s")
print(f"Período orbital: {T / 31536000:.3e} años")
print(f"Excentricidad orbital: {e:.3f}")
```
